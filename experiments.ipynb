```python
from langchain.docstore.document import Document

docs = [Document(page_content=row['text']) for _, row in df.iterrows()]

docs
```

```python
from langchain_qdrant import Qdrant
from langchain_community.embeddings.huggingface import HuggingFaceEmbeddings



embeddings = HuggingFaceEmbeddings(
    model_name="DeepPavlov/distilrubert-tiny-cased-conversational-v1"
)

qdrant = Qdrant.from_documents(
            docs,
            embeddings,
            location=":memory:",
            collection_name="my_documents",
)
       
qdrant_retriever = qdrant.as_retriever(search_kwargs={"k": 10}, verbose=True)
```

```python
import re

def contains_russian(text):
    # Define a regular expression pattern to match any Russian character
    russian_pattern = r'[–∞-—è–ê-–Ø—ë–Å]'
    
    # Search for the pattern in the text
    if re.search(russian_pattern, text):
        return True
    else:
        return False

# Example usage
examples = [
    "This is a normal text.",
    "–≠—Ç–æ —Ç–µ–∫—Å—Ç –Ω–∞ —Ä—É—Å—Å–∫–æ–º —è–∑—ã–∫–µ.",
    "Mixed text with English and —Ä—É—Å—Å–∫–∏–π.",
    "123456",
    "–¢–µ–∫—Å—Ç –±–µ–∑ –∞–Ω–≥–ª–∏–π—Å–∫–∏—Ö –±—É–∫–≤.",
]

for example in examples:
    print(f"'{example}' contains Russian characters: {contains_russian(example)}")
```

```python
import numpy as np
import pandas as pd
import json
import requests
from tqdm import tqdm_notebook
from openai import OpenAI
import openai

openai.api_key = 'token'

with open("data/texts.csv", "r") as f:
  all_text = f.read()


share_dict = {
    .25: {
        'col': [],
        'name': '25%_changing'
    },
    .5: {
        'col': [],
        'name': '50%_changing'
    },
    .75: {
        'col': [],
        'name': '75%_changing'
    },
    1.0: {
        'col': [],
        'name': '100%_changing'
    },
}

for text_ind in range(len(all_text)):
  split_text = all_text.iloc[text_ind, 0].split('.')
  cum_share = {key: len(key) / sum([len(sent) for sent in split_text]) for key in split_text}
  for ind in range(len(cum_share)):
    cur_key = list(cum_share.keys())[ind]
    if ind != 0:
      prev_key = list(cum_share.keys())[ind - 1]
      cum_share[cur_key] += cum_share[prev_key]

  for share in [.25, .5, .75, 1]:
    tmp_filter = dict(filter(lambda item: item[1] > share, cum_share.items()))
    cur_text = '.'.join(list(cum_share.keys())[:len(cum_share) - len(tmp_filter)])

    prompt = f"""
                Paraphrase the following text directly: {cur_text}

                You cannot change the meaning of the original sentence.
                You cannot write introductory words or immediately
                paraphrased text, for example:

                Prompt: "Rephrase the following without any introductory words:
                The quick brown fox jumps over the lazy dog."
                Response: "The swift brown fox leaps over the lazy dog."
            """
    
    response = openai.chat.completions.create(
          model='gpt-3.5-turbo',
          messages=[
              {"role": "user", "content": f"Paraphrase the following text: {prompt}"}
          ]
      )
    # Extract the paraphrased text from the response
    paraphrased_text = response.choices[0].message.content.strip()

    share_dict[share]['col'].append(paraphrased_text + '.'.join(list(cum_share.keys())[len(cum_share) - len(tmp_filter):]))
```
